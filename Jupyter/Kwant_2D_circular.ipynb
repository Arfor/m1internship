```python
import holoviews as hv
hv.extension('matplotlib', logo=False) #BOKEH DOESNT SUPPORT LATEX IN LABELS
from holoviews import opts
hv.output(fig='svg')
import kwant
import numpy as np
import pandas as pd
import tinyarray as tiny
from types import SimpleNamespace
```

```python
sys = kwant.Builder()
param = SimpleNamespace(r=15)
sqlat = kwant.lattice.square()
def disk(position):
    x,y = position
    return x**2 + y**2 < param.r**2
sys[sqlat.shape(disk,(0,0))] = 1 #onsite energy 
sys[sqlat.neighbors(1)] = -1 #hopping term
plot = kwant.plot(sys) #plot the shape of the lattice
```

```python
type(plot)
hv.render(plot)
```
