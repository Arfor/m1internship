# 6.3. Creating interactive Web visualizations with Bokeh and HoloViews

```python
import numpy as np
from numpy import linalg as LA
import holoviews as hv
from holoviews import opts
hv.extension('bokeh', 'matplotlib')
```

```python
n=5
r=(np.random.rand(n,n)-0.5)*2
H0= (r+r.T)/2
r=(np.random.rand(n,n)-0.5)*2
H1= (r+r.T)/2

a = np.linspace(0,1,15)

for x in a:
    H_cont = x*H1 + (1-x)*H0
    
eVal0, eVec0 = LA.eig(H0) #returns eigenvectors as column vectors
```

```python
hv.Points(np.full(n,0),eVal0)
```

```python
data = np.random.rand(100, 100, 10)
```

```python
ds = hv.Dataset((np.arange(10),
                 np.linspace(0., 1., 100),
                 np.linspace(0., 1., 100),
                 data),
                kdims=['time', 'y', 'x'],
                vdims=['z'])
```

```python
ds
```

```python
%opts Image(cmap='viridis')
ds.to(hv.Image, ['x', 'y']).hist()
```
