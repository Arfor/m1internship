### Magnetic texture of Skyrmion

```python
import numpy as np
# import kwant
from types import SimpleNamespace
import holoviews as hv
from holoviews import opts
from matplotlib import pyplot as plt
hv.extension('matplotlib', logo=False) #BOKEH DOESNT SUPPORT LATEX IN LABELS
hv.output(fig='svg')
```

```python
def magn_skyrmion(position,params):
    x,y = position
    t = np.arctan2(y,x)
    q = params.aziwinding
    p = params.radiwinding
    R = params.radius
    r = np.sqrt(x**2 + y**2)
    B = [np.sin(np.pi*p*(r/R))*np.cos(q*t), np.sin(np.pi*p*(r/R))*np.sin(q*t), np.cos(np.pi*p*(r/R))]
    return B
```

```python
params = SimpleNamespace(t=1, mu=-0.1, j=0.02, delta=0.2, B_0=0.1, aziwinding=1, radiwinding=2,radius=15)
def texture(params, q=1,p=1):
    init = []
    params.aziwinding=q
    params.radiwinding=p
    r= params.radius
    x = np.arange(-r,r,1)
    y = np.arange(-r,r,1)
    for i in x:
        for j in y:
            init.append(magn_skyrmion((i,j),params))
    return np.reshape(np.asarray(init),(len(x),len(y),3), order='F')
```
