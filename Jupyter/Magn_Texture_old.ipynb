### Magnetic texture of Skyrmion

```python
import numpy as np
# import kwant
from types import SimpleNamespace
import holoviews as hv
from holoviews import opts
from matplotlib import pyplot as plt
hv.extension('matplotlib', logo=False) #BOKEH DOESNT SUPPORT LATEX IN LABELS
hv.output(fig='svg')
```

```python
params = SimpleNamespace(t=1, mu=-0.1, j=0.02, delta=0.2, B_0=0.1, aziwinding=1, radiwinding=1,radius=20)
def magn_skyrmion(position,params):
    x,y = position
    theta = np.arctan2(x,y)
    q = params.aziwinding
    p = params.radiwinding
    R = params.radius
    r = np.sqrt(x**2 + y**2)
    B = [np.sin(np.pi*p*(r/R))*np.cos(q*theta), np.sin(np.pi*p*(r/R))*np.sin(q*theta), np.cos(np.pi*p*(r/R))]
    return B

def texture(params, q=1,p=1):
    init = []
    params.aziwinding=q
    params.radiwinding=p
    r= params.radius
    x = np.arange(-r,r,1)
    y = np.arange(-r,r,1)
    for i in x:
        for j in y:
            init.append(magn_skyrmion((i,j),params))
    return np.reshape(np.asarray(init),(len(x),len(y),3))
```

```python
r= params.radius
hv.HoloMap({(i,q,p): hv.Image(texture(params,q=q,p=p)[:,:,i], bounds=(r,r,-r,-r)).opts(colorbar=True, cmap='inferno') for i in range(3) \
            for q in range(1,4) for p in range(1,4)}, kdims=['Component','Azimuthal Winding Number', 'Radial Winding Number'])
```
